```python
import seaborn
import pandas as pd
import seaborn as sns
import datetime 

sns.set_theme(style="darkgrid")
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
df = pd.read_excel('covid_19_data_portal_data.xlsx')
df['Date'] = df['Period']
df["Date"]= pd.to_datetime(df["Date"])

df.head()
```

```python
econ_confid = df[df.ResourceID =='CPCON1'].set_index('Period')
no_cases_tests = df[(df.ResourceID =='CPCOV2') | (df.ResourceID =='CPCOV1')].set_index('Period')
no_vax_cases = df[(df.ResourceID =='CPCOV9') | (df.ResourceID =='CPCOV2')].set_index('Period')
case_status = df[(df.ResourceID =='CPCOV4')].set_index('Period')

case_status.head()
#case_status.dtypes
```

```python
e_confid = sns.lineplot(x="Period", y="Value",
             data=econ_confid)
for ind, label in enumerate(e_confid.get_xticklabels()):
    if ind % 15 == 0:  # every 10th label is kept
        label.set_visible(True)
        label.set_rotation(90)
    else:
        label.set_visible(False)
```

```python
no_cases_tests_plot = sns.lineplot(x="Period", y="Value", hue="Measure",
             data=no_cases_tests)
for ind, label in enumerate(no_cases_tests_plot.get_xticklabels()):
    if ind % 15 == 0:  # every 10th label is kept
        label.set_visible(True)
        label.set_rotation(90)
    else:
        label.set_visible(False)
```

```python
no_vax_cases_plot = sns.lineplot(x="Period", y="Value", hue="Measure",
             data=no_vax_cases)
for ind, label in enumerate(no_vax_cases_plot.get_xticklabels()):
    if ind % 15 == 0:  # every 10th label is kept
        label.set_visible(True)
        label.set_rotation(90)
    else:
        label.set_visible(False)
```

```python
case_status_plot = sns.lineplot(x="Period", y="Value", hue="Measure",
             data=case_status)
for ind, label in enumerate(case_status_plot.get_xticklabels()):
    if ind % 15 == 0:  # every 10th label is kept
        label.set_visible(True)
        label.set_rotation(90)
    else:
        label.set_visible(False)
```
